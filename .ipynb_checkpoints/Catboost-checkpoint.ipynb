```python
def get_roc_eval(y_test, pred=None, pred_proba=None): 
    confusion = confusion_matrix(y_test, pred) 
    accuracy = accuracy_score(y_test, pred) 
    precision = precision_score(y_test, pred) 
    recall = recall_score(y_test, pred) 
    f1 = f1_score(y_test, pred) 
    # ROC AUC 
    roc_auc = roc_auc_score(y_test, pred_proba) 
    return roc_auc
```

```python
from sklearn.metrics import *

def get_clf_eval(y_test, pred=None, pred_proba=None): 
    confusion = confusion_matrix(y_test, pred) 
    accuracy = accuracy_score(y_test, pred) 
    precision = precision_score(y_test, pred) 
    recall = recall_score(y_test, pred) 
    f1 = f1_score(y_test, pred) 
    # ROC AUC 
    roc_auc = roc_auc_score(y_test, pred_proba) 
    
    print('Confusion Matrix') 
    print(confusion) 
    print('accuracy: {0:.4f}, precision: {1:.4f}, recall: {2:.4f}, f1: {3:.4f}, roc_auc: {4:.4f}'.format( accuracy, precision, recall, f1, roc_auc))
```

### 데이터 엔지니어링

```python
train  = pd.read_csv('data/train.csv')
test  = pd.read_csv('data/test.csv')
```

```python
tipi_list = [f'TIPI{i + 1}' for i in range(10)]
drop_list = ['country','index']

for i in range(len(tipi_list) // 2):
    if i % 2 == 0:
        po_ti=tipi_list[i]
        ne_ti=tipi_list[i + 5]

    elif i % 2 == 1:
        po_ti=tipi_list[i + 5]
        ne_ti=tipi_list[i]

    train[ne_ti]=train[ne_ti].apply(lambda x : 6-x)
    train[f'gen_TIPI{i + 1}']=(train[po_ti]+train[ne_ti])/2
    test[ne_ti]=test[ne_ti].apply(lambda x : 6-x)
    test[f'gen_TIPI{i + 1}']=(test[po_ti]+test[ne_ti])/2
```

```python
# VCL scoring
vcl_list = [f'VCL{i + 1}' for i in range(16)]

train_vcl_score_temp = train[vcl_list[0]]
test_vcl_score_temp = test[vcl_list[0]]
for i in vcl_list[1:]:
    if i == 'VCL6' or i == 'VCL9' or i == 'VCL12':
        train_vcl_score_temp -= train[i] * 4
        test_vcl_score_temp -= test[i] * 4
    else:
        train_vcl_score_temp += train[i]
        test_vcl_score_temp += test[i]

train['VCL_score'] = train_vcl_score_temp
test['VCL_score'] = test_vcl_score_temp
```

```python
train = train.drop(train[train.familysize > 50].index)
train = train.drop(train[train.age > 120].index)
train_y = train['nerdiness']
train_x = train.drop(drop_list + tpip_list + ['nerdiness'], axis=1)
test = test.drop(drop_list + tpip_list, axis=1)
train_x.reset_index(drop=True,inplace=True)
train_x
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score

accuracy_history = []
str_kf = StratifiedKFold(n_splits = 10, shuffle = True, random_state = 50)

final_result=[]
for train_index, test_index in str_kf.split(train_x,train_y):
    X_train, X_test = train_x.iloc[train_index], train_x.iloc[test_index]
    y_train, y_test = train_y.iloc[train_index], train_y.iloc[test_index]
    
    lgbm_wrapper = LGBMClassifier(n_estimators=5000)
    lgbm_wrapper.fit(X_train, y_train, early_stopping_rounds=2000, eval_metric='logloss', eval_set=evals, verbose=False)    
    preds = lgbm_wrapper.predict(X_test)
    pred_proba = lgbm_wrapper.predict_proba(X_test)[:,1]
    get_roc_eval(y_test, preds, pred_proba)
    sub_pred_proba =lgbm_wrapper.predict_proba(test)[:,1]
    final_result.append(sub_pred_proba)
    accuracy_history.append(get_roc_eval(y_test, preds, pred_proba))
    
print("각 분할의 정확도 기록 :", accuracy_history)
print("평균 정확도 :", np.mean(accuracy_history))
```
