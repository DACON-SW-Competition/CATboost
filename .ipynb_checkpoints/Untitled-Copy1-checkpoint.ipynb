### TIPI 점수 계산

```python
train_data  = pd.read_csv('data/train.csv')
test_data  = pd.read_csv('data/test.csv')
```

```python
# TIPI 전처리
drop_list = ['country','index']
tpip_list=['TIPI1', 'TIPI2', 'TIPI3', 'TIPI4','TIPI5', 'TIPI6', 'TIPI7', 'TIPI8', 'TIPI9', 'TIPI10']
for i in range(len(tpip_list)):
    if i ==5:
        break
    po_ti=tpip_list[i]
    ne_to=tpip_list[i+5]
    train_data[ne_to]=train_data[ne_to].apply(lambda x : 6-x)
    train_data[f'gen_TIPI{i}']=(train_data[po_ti]+train_data[ne_to])/2

    test_data[ne_to]=test_data[ne_to].apply(lambda x : 6-x)
    test_data[f'gen_TIPI{i}']=(test_data[po_ti]+test_data[ne_to])/2
    

train_data = train_data.drop(train_data[train_data.familysize > 50].index)
train_data = train_data.drop(train_data[train_data.age > 120].index)


train_y = train_data['nerdiness']
train_x = train_data.drop(drop_list + tpip_list + ['nerdiness'], axis=1)
test_x = test_data.drop(drop_list + tpip_list, axis=1)
```

```python
lgbm_wrapper = LGBMClassifier(n_estimators=5000)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(train_x, train_y, test_size=0.1,random_state=7942)
```

```python
evals = [(X_test, y_test)]
lgbm_wrapper.fit(X_train, y_train, early_stopping_rounds=1000, eval_metric='logloss', eval_set=evals, verbose=True)
```

```python
preds = lgbm_wrapper.predict(X_test)
```

```python
pred_proba = lgbm_wrapper.predict_proba(X_test)[:,1]
```

```python
from sklearn.metrics import *

def get_clf_eval(y_test, pred=None, pred_proba=None): 
    confusion = confusion_matrix(y_test, pred) 
    accuracy = accuracy_score(y_test, pred) 
    precision = precision_score(y_test, pred) 
    recall = recall_score(y_test, pred) 
    f1 = f1_score(y_test, pred) 
    # ROC AUC 
    roc_auc = roc_auc_score(y_test, pred_proba) 
    
    print('Confusion Matrix') 
    print(confusion) 
    print('accuracy: {0:.4f}, precision: {1:.4f}, recall: {2:.4f}, f1: {3:.4f}, roc_auc: {4:.4f}'.format( accuracy, precision, recall, f1, roc_auc))
```

```python
get_clf_eval(y_test, preds, pred_proba)
```
